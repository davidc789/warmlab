# Simple

```python
import igraph as ig
import numpy as np
from warmlab import warm
import matplotlib.pyplot as plt

np.random.seed(10)
```

```python
edges = [
    (0, 1), (1, 2), (2, 3), (3, 0), (0, 4), (0, 5)
]

graph = ig.Graph(edges)
model = warm.WarmModel(
    is_graph=True,
    graph=graph
)
solution = warm.solve(model)
x_opt = np.array(solution.x_opt)
x_opt
```

```python
fig, ax = plt.subplots()
ig.plot(graph, target=ax, edge_width=x_opt * 20)
```
